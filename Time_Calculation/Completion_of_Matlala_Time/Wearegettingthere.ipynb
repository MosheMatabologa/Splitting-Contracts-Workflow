```python
import pandas as pd
import datetime as datetime
```

```python
file = r'C:\Users\Q624157\Desktop\Time_Calculation\Clocking Data.XLSX'
```

```python
hours_in_day = 48
```

```python
df = pd.read_excel(file, sheet_name="Sheet1")
```

```python
#Calculate total time clocked for the day
df['Clocking Totals (Day)'] = df['Clock Out Time'] - df['Clock In Time']
```

```python
# Convert the column to string type before applying .str methods
df['Time Event In/Out Time'] = df['Time Event In/Out Time'].astype(str)

# Now apply the .str.replace() method
df['Clock Out Time'] = pd.to_timedelta(df['Time Event In/Out Time'].str.replace('24:00', '00:00:00'), errors='coerce')
df['Clock In Time'] = pd.to_timedelta(df['Time Event In/Out Time'].str.replace('24:00', '00:00:00'), errors='coerce')
```

```python
df.head()
```

```python
#Calculate total time clocked for the day
df['Clocking Totals (Day)'] = df['Clock Out Time'] - df['Clock In Time']
```

```python
df['Time Event Date'] = pd.to_datetime(df['Time Event Date'])
```

```python
unique_days = df['Time Event Date'].dt.date.unique()
```

```python
# Iterate over the unique days
for day in unique_days:
    # Filter the DataFrame for rows with the same day
    day_df = df[df['Time Event Date'].dt.date == day]
    
    # Create a new datetime column by combining date and time
    day_df['DateTime'] = pd.to_datetime(day_df['Time Event Date'].dt.date.astype(str) + ' ' + day_df['Time Event In/Out Time'].astype(str))
    
    # Calculate the total hours between 14:00 and 18:00
    hours_14_to_18 = ((day_df['DateTime'].dt.hour >= 14) & (day_df['DateTime'].dt.hour < 18)).sum() / 2
    
    # Write the result to column J
    df.loc[df['Time Event Date'].dt.date == day, 'Column J'] = hours_14_to_18
    
    # Calculate the total hours between 06:00 and 18:00
    hours_06_to_18 = ((day_df['DateTime'].dt.hour >= 6) & (day_df['DateTime'].dt.hour < 18)).sum() / 2
    
    # Write the result to column K
    df.loc[df['Time Event Date'].dt.date == day, 'Column K'] = hours_06_to_18
    
    # Print the result
    print(f"Total hours for {day}: {hours_14_to_18} hours between 14:00 and 18:00, {hours_06_to_18} hours between 06:00 and 18:00")
```
